```python
var = df_diamantes['color'].nunique()
var = 30//var
sample2 = df_diamantes.groupby('color').sample(n=var)
sample2.shape[0]
```

```python
def amostra_grupo(grupo, quantidade):
    amostra = grupo.sample(quantidade)
    return amostra

grupos = df_diamantes['color'].value_counts()
grupos = grupos/len(df_diamantes)
# agora grupos é a fração correta pra cada um dos estratos
grupos = grupos * 30
qnt_por_grupo = grupos.round()
colors_grupos = df_diamantes['color'].unique()
colors_grupos = colors_grupos
colors_grupos = qnt_por_grupo.index.tolist()

print(qnt_por_grupo)
print(colors_grupos)

amostra_final = pd.DataFrame()

for i in range(qnt_por_grupo.shape[0]):
    df_temp = amostra_grupo(df_diamantes[df_diamantes['color'] == colors_grupos[i]],int(qnt_por_grupo[i]))
    amostra_final = pd.concat([amostra_final,df_temp],ignore_index=True)

amostra_final

    
```
